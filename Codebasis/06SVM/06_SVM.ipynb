```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dataset = pd.DataFrame(digits.data)
dataset.head()
```

```python
dataset['target'] = digits.target
dataset.head()
```

```python
X=dataset.drop('target',axis='columns')
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

**Using RBF kernal**

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
model=SVC()
```

```python
rbf_model.fit(X_train, y_train)
```

```python
rbf_model.score(X_test, y_test)
```

**Using Linear Kernal**

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
linear_model.score(X_test,y_test)
```

**Using Poly, Sigmoid and precomputer Kernal**

```python
poly_kernal=SVC(kernel='poly')
sig_kernal=SVC(kernel='sigmoid')
pre_kernal=SVC(kernel='precomputed')
```

```python
poly_kernal.fit(X_train, y_train)
sig_kernal.fit(X_train, y_train)
# pre_kernal.fit(X_train, y_train)
```

```python
poly_kernal.score(X_test, y_test)
```

```python
sig_kernal.score(X_test, y_test)
```
